## Test Codes (Time Issues)

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

def check_dates(datetime_str, desired_format, replace_hour):
    try: 
        parsed_dt = datetime.strptime(datetime_str, desired_format)
        return datetime.strftime(parsed_dt, desired_format)
    except ValueError: 
        # If datetime string doesn't match desired format or doesn't start with '29', return as is
        if datetime_str.startswith("29"):
            # Check if the datetime string includes milliseconds
            if '.' in datetime_str:
                # Extract milliseconds
                milliseconds = datetime_str.split(".")[-1]
                # Parse the time part without milliseconds
                time_part = datetime.strptime(datetime_str.split(".")[0], "%H:%M:%S")
                # Replace the hour with replace_hour
                modified_dt = time_part.replace(hour=int(replace_hour))
                # Append milliseconds
                modified_dt = modified_dt.replace(microsecond=int(milliseconds) * 1000)  # Multiply milliseconds by 1000 to convert to microseconds
                return datetime.strftime(modified_dt, "%H:%M:%S.%f")
            else:
                # No milliseconds, so parse as usual and replace the hour
                time_part = datetime.strptime(datetime_str, "%H:%M:%S")
                modified_dt = time_part.replace(hour=int(replace_hour))
                return datetime.strftime(modified_dt, "%H:%M:%S")
        else:
            return datetime_str
    

def replace_hours_in_column(column, desired_format, replace_hour):
    modified_column = []
    for dt_str in column: 
        modified_column.append(check_dates(dt_str, desired_format, replace_hour))
    return modified_column


# Location of files to be processed
data_location = r"C:\Users\pawoo\OneDrive\Documents\Coastal Hazards Lab Stuff\Research Project Items and Python\SDF\Wet-Dry Calibration\Data Logs and Images\PT_23 20 Micron Half Cavity + Stone Calibration"
# Define full path to the file
file_path = os.path.join(data_location, "dataLog00018.TXT")

# Read text file into pandas DataFrame 
datadf = pd.read_csv(file_path)

desired_format = "%H:%M:%S.%f"
replace_hour = '12'
datadf['rtcTime'] = replace_hours_in_column(datadf['rtcTime'], desired_format, replace_hour)

print(datadf)

```
