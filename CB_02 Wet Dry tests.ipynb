# CB_02 Wet Dry Tests

```python
# Imports 

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

## Prior to March 5th - 10 Micron Stone Filter + Unknown Gel 
### Note that water was found to be trapped behind the gel 

```python

# Load the data
total_data = pd.read_csv('CB_02_20240221_20240305.csv')

# Convert 'timestamp (UTC)' to datetime
total_data['timestamp (UTC)'] = pd.to_datetime(total_data['timestamp (UTC)'])

# Define the weekly intervals
start1 = datetime(2024, 2, 21)
end1 = datetime(2024, 2, 28)
start2 = datetime(2024, 2, 28)
end2 = datetime(2024, 3, 5)

# Plot the data for the first 1-week interval
plt.figure(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Water Level (ft. NAVD88)')
plt.title('10 Micron Filter with Unknown Gel Amount (Feb. 21 - Feb. 28)')
plt.xlim([start1, end1])
plt.ylim(.2, 3.5)
plt.axhline(y=0.75, color='coral', linestyle='--', label='Sensor Elevation')
plt.axhline(y=2.39, color='teal', linestyle='--', label='Road Elevation')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(pd.date_range(start=start1, end=end1, freq='D'))
plt.gcf().autofmt_xdate()
plt.legend(fontsize='small', loc='upper left')
interval1_data = total_data[(total_data['timestamp (UTC)'] >= start1) & (total_data['timestamp (UTC)'] <= end1)]
plt.plot(interval1_data['timestamp (UTC)'], interval1_data['water level (ft NAVD88)'], label='Water Level')
plt.legend()
plt.show()

# Plot the data for the second 1-week interval
plt.figure(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Water Level (ft. NAVD88)')
plt.title('10 Micron Filter with Unknown Gel Amount (Feb. 28 - Mar. 5)')
plt.xlim([start2, end2])
plt.ylim(.2, 3.5)
plt.axhline(y=0.75, color='coral', linestyle='--', label='Sensor Elevation')
plt.axhline(y=2.39, color='teal', linestyle='--', label='Road Elevation')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(pd.date_range(start=start2, end=end2, freq='D'))
plt.gcf().autofmt_xdate()
plt.legend(fontsize='small', loc='upper left')
interval2_data = total_data[(total_data['timestamp (UTC)'] >= start2) & (total_data['timestamp (UTC)'] <= end2)]
plt.plot(interval2_data['timestamp (UTC)'], interval2_data['water level (ft NAVD88)'], label='Water Level')
plt.legend()
plt.show()
```
